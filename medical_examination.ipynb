```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
df['overweight'] = np.where((df['weight'] / ((df['height'] / 100) **2)) > 25, 1, 0)
df
```

```python
df[df['overweight'] == 1]
```

```python
df['cholesterol'] = np.where(df['cholesterol'] == 1, 0 ,1)
```

```python
df['gluc'] = np.where(df['gluc'] == 1, 0 ,1)
```

```python
df
```

```python
a = pd.melt(df, value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'], id_vars=['cardio'])
a
```

```python
grouped_df = a.groupby(['cardio', 'variable', 'value']).size().reset_index(name='total')
grouped_df
```

```python
catplot = sns.catplot(data=grouped_df, x='variable', y='total', hue='value', col='cardio', kind='bar') 
catplot.fig.set_size_inches(15,8)
catplot
```

```python
fig = catplot.fig
fig
fig.savefig('catplot.png')
```

```python
df2 = df[(df['ap_lo'] <= df['ap_hi']) & 
         df['height'].between(df['height'].quantile(0.025), df['height'].quantile(0.975), inclusive='both') & 
         df['weight'].between(df['weight'].quantile(0.025), df['weight'].quantile(0.975), inclusive='both')].copy()
df2
```

```python
mask = np.triu(df2.corr())
fig, ax = plt.subplots(figsize=(15,12))
ax.set_facecolor("white")
sns.set(font_scale=1.3)
chart = sns.heatmap(df2.corr(), annot=True, mask=mask, fmt=".1f", center = 0.0, 
            cbar_kws={"shrink":0.5, "ticks":[-0.08, 0.00, 0.08, 0.16, 0.24]}, 
            linewidths=1, vmax=0.30, vmin=-0.14
)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
fig.savefig('heatmap.png')
```
